```python
# Generating an artificial multi-label dataset

from sklearn.datasets import make_multilabel_classification

# this will generate a random multi-label dataset

X, y = make_multilabel_classification(sparse = True, n_labels = 20,return_indicator = 'sparse', allow_unlabeled = False)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,random_state=42)
 
```

```python
# using binary relevance
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB

# initiase binary relevance multi-label classifier
# with a gausian naive bayes base classifier

classifier = BinaryRelevance(GaussianNB())

# train
classifier.fit(x_train, y_train)

# predict 
predictions = classifier.predict(x_test)
```
